# **1. Introduction and Description of the Dataset**
This dataset contains information about the top 100 nightclubs in 2024. The columns in the dataset are as follows:

*   **Rank**: Position in the ranking.
*   **Club_name**: Name of the club.
*   **Location**: City and country where the club is located.
*   **Capacity**: Capacity of the club (number of people it can accomodate).
*   **Website**: Official website of the club.
*   **Text**: Additional description about the club.





# **2. Loading and Exploring the Dataset**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/top_100_clubs.csv') # insert real path in order to execute correctly

# Display the first few rows of the dataset
print(df.head())

# Display basic information about the dataset
print(df.info())
```

# **3. Data Cleaning and Preparation**

```python
# Clean the 'capacity' column
df['capacity'] = df['capacity'].replace({
    '8,000+': '8000',
    '3,000 (Nightclub), 1,600 (Dayclub)taogroup.com/venues/marquee-las-vegas': '3000',
    '2,000+': '2000'
})
df['capacity'] = df['capacity'].str.replace(',', '').astype(int)

# Extract the country from the 'location' column
df['Country'] = df['location'].apply(lambda x: x.split(", ")[-1])
```

# **4. Univariate Analysis**

## **Distribution of Clubs by Country**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of clubs per country
country_counts = df['Country'].value_counts()

# Visualize the distribution of clubs by country
plt.figure(figsize=(12, 8))
sns.countplot(y='Country', data=df, order=country_counts.index)
plt.title('Distribution of Clubs by Country')
plt.xlabel('Number of Clubs')
plt.ylabel('Country')
plt.show()
```

## **Distribution of Club Capacities**


```python
# Visualize the distribution of club capacities
plt.figure(figsize=(12, 6))
sns.histplot(df['capacity'], bins=20, kde=True)
plt.title('Distribution of Club Capacities')
plt.xlabel('Capacity')
plt.ylabel('Frequency')
plt.show()
```

# **5. Conclusions**

1.   **Geographical Distribution**: The majority of the top nightclubs are located in Europe, with significant clusters in countries like Spain and the United Kingdom.
2.   **Capacity of Clubs**: Club capacities vary widely, ranging from less than 500 to over 10,000. Larger capacities are often found in prominent nightlife hubs.
3.   **Country-Capacity Relationship**: Some countries, particularly those with renowned nightlife scenes, tend to have clubs with higher capacities.
